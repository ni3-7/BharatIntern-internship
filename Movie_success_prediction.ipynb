#  Using Machine learning

#### The code imports essential modules from scikit-learn (sklearn) for machine learning tasks, including data splitting, feature scaling (StandardScaler), training a random forest classifier (RandomForestClassifier), and evaluating model performance using metrics like accuracy, classification report, confusion matrix, and ROC curve (roc_curve).

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve
```

```python
# Split the dataset into training and testing sets using 20% for testing and a random state of 10.
x_train, x_test,y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
# Printing the shapes of the original feature set (x), training set (x_train), and testing set (x_test).
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
# Standardizing the features in the training set (x_train) using the StandardScaler and applying the same transformation to the testing set (x_test).
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# Initializing a Random Forest Classifier with 60 decision trees (n_estimators=60).
rf = RandomForestClassifier(n_estimators=60)
```

```python
# Fitting the Random Forest Classifier to the training data and making predictions on both the training set (y_pred_rf_train) and testing set (y_pred_rf_test).
rf.fit(x_train, y_train)
y_pred_rf_test = rf.predict(x_test)
y_pred_rf_train = rf.predict(x_train)
```

```python
# Calculating the accuracy scores for the Random Forest Classifier on the testing set (testacc_rf) and training set (trainacc_rf).
testacc_rf = accuracy_score(y_pred_rf_test, y_test)
trainacc_rf = accuracy_score(y_pred_rf_train, y_train)
```

```python
# Printing the training accuracy and testing accuracy of the Random Forest Classifier.
print("Training Accuracy of Random Forest: ", trainacc_rf)
print("Testing Accuracy of Random Forest: ", testacc_rf)
```

```python
# Creating a heatmap to visualize the confusion matrix of the Random Forest Classifier on the testing set.
sns.heatmap(confusion_matrix(y_test,y_pred_rf_test), fmt = 'd',annot=True, cmap='magma')
```

```python
# Printing the classification report of the Random Forest Classifier on the testing set.
print(classification_report(y_test,y_pred_rf_test))
```
